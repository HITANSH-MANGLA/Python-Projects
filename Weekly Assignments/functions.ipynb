**WAP in python to create a function which takes input of all datatype and return list of integers**

```python
#creating a function which takes list of all datatypes as input but only returns list of integers
def hit(a):
  """this function is used to return the list of integers"""
  b=[]
  for i in a:
    if type(i)==int:
      b.append(i)
  return b


a =[2,3.8,9]
print(hit(a))
```

```python
#creating a function which inputs a number and create a list from 1 to till then number
def list_four(a):
  """this function is used to create a list from1 to the input number if and only if that is divisible by four """
  
  l=[]
  for i in range(1,a):
    if i%4==0:
      l.append(i)
  return l


list_four(int(input("enter the number")))

```

```python
#creating a function which takes input of list and returns the unique elemnt from it
def unique_list(l):
  """this funtion is used to return the unique element from the given list"""
  
  for i in l:
    if l.count(i) == 1:
      print(i)

l1 = [3, 3, 3, 7, 3, 85,73]
unique_list(l1)
```

```python
#creating a function which takes inpu of string and replaces all the vowels of that vowel with a specific vowel
def replace_vowel(k):
  """this function is used to replace the vowels in a given string to another specfic alphabet"""
  q=k
  v="aeiou"
  for i in k:
    if i in v:
     q= q.replace(i,'u')
  return q


k='hitansh'
replace_vowel(k)
```
