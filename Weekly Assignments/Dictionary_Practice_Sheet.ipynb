**QUESTION:- WAP IN PYTHON TO SUM ALL THE ITEMS IN DICTIONARY**


```python
#sum of all the items in a dictionary
a={}
s=0
k=0
m=0
size=int(input("enter the size of dictionary"))
for i in range(0,size):
  key=int(input("enter key"))
  value=int(input("enter value"))
  a[key]=value
for j in a.keys():
  s=s+j
for l in a.values():
  k=k+l
m=s+k
print("the sum of all the items of dictionary",m)
```

**QUESTION:-WAP IN PYHTON TO PRINT A DICTIONARY WHERE KEYS ARE BETWEEN 1 AND 15 AND VALUES ARE THE SQUARE OF KEYS**

**QUESTION:-WAP IN PYTHON TO MAP TWO LISTS INTO DICTIONARY**

```python
#maping two list in the format of dictionary
a=[1,2,3,4,5]
b=[6,7,8,9,10]
d=dict(zip(a,b))
d
```

**QUESTION:- WAP IN PYTHON TO MAP TWO LISTS INTO DICTIONARY WITHOUT ZIP FUNCTION**

```python
#maping two lists into dictionary without zip function 
a=[1,2,3,4,5]
b=[6,7,8,9,10]
d=dict(map(lambda a,b:{a,b},a,b))
d
```
