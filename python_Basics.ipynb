```python
#calculating lcm and hcf of number 
a=int(input("enter the first number"))
b=int(input("enter the second number"))
if(a>b):
  min=b
else:
  min=a
for i in range(1,min):
  if(a%i==0 and b%i==0):
    hcf=i

print("the hcf of these two numbers is",hcf)
lcm=a*b//hcf
print("the lcm of the number is",lcm)
```

```python
#pattern printing
# *
# **
# ***
# ****
# *****
n=int(input("enter the value"))
i=1
j=1
for i in range(1,n+1):
  for j in range(1,i):
    print("*",end=" ")
  print("\n")
```
